# Check whether a distribution is normal or not
using QQ Plot - Quantile Quantile Plot

```python
# importing libaries
import scipy.stats as stat
import matplotlib.pyplot as plt
import pylab
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# making a function which plots the distribution of a feature in dataframe
def plot_data(df, feature) :
    plt.figure(figsize=(10, 6))
    # left -> Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[feature], kde=True)
    # right -> QQ Plot
    plt.subplot(1, 2, 2)
    stat.probplot(df[feature], dist='norm', plot=pylab)
    plt.show()
```

```python
plot_data(df, 'sepal_width')
# follows normal distribution
```

```python
"""
With the help of QQ Plot we will be able to understand whether a is distribution is normal or not
On left hand side, we have a smoothened histogram, which looks like a bell curve
On right hand side, we have a QQ plot, on X axis we have a normal distribution which was initialized
                                       on Y axis we have feature value
                    if most of the points fall in one straight line, then this is called as normal distribution.
"""
```

```python
plot_data(df, 'petal_width')
# do not follow normal distribution since most of points do not lie on red line 
```

```python
plot_data(df, 'sepal_length')
# follows a normal distribution
```

```python
plot_data(df, 'petal_length')
# Do not follow normal distribution
```
