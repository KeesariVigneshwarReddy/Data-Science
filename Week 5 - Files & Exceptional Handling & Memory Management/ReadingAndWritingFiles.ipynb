```python
# storing in json file
data = {
           "name" : "vignesh",
           "email" : "vignesh@gmail.com",
           "phone number" : 9630258741,
           "subject" : ["Java", "Data Science"]
       }
```

```python
# write operation with json
import json
with open("test1.json", 'w') as f :
    json.dump(data, f)
```

```python
# read operation with json
with open("test1.json", 'r') as f :
    data1 = json.load(f)
```

```python
# list of list
# storing in csv file (comma seperated values)
data = [
           ["name", "email", "phone number"],
           ["Vignesh", "vignesh@gmail.com", "7401258963"],
           ["Rohan", "rohan@gmail.com", "9630258741"]
       ]
```

```python
# write operation with csv
import csv
with open("test3.csv", 'w') as f :
    w = csv.writer(f)
    for i in data :
        w.writerow(i)
```

```python
# read operation with csv
with open("test3.csv", 'r') as f :
    read = csv.reader(f)
    for i in read :
        print(i)
```
