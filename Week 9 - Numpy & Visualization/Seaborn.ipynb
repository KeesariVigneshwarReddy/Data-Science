```python
import seaborn as sns
```

```python
# loading a sample dataset
# data frame
sns.load_dataset('iris')
```

```python
iris = sns.load_dataset('iris')
```

```python
# sepal length vs sepal width
sns.scatterplot(x = iris.sepal_length, y = iris.sepal_width)
```

```python
sns.lineplot(x = iris.sepal_length, y = iris.sepal_width)
```

```python
# sepal length vs prtal length
sns.scatterplot(x = iris.sepal_length, y = iris.petal_length)
```

```python
# histogram
sns.displot(iris['sepal_length'])
```

```python
sns.load_dataset('tips')
```

```python
tips = sns.load_dataset('tips')
```

```python
# trend between total_bill vs tip
sns.scatterplot(x=tips.total_bill, y = tips.tip)
```

```python
tips.head()
```

```python
# determine count of smoker and non smoker
tips['smoker'].value_counts()
```

```python
# representing smoker or non smoker in same graph
# color coding
sns.relplot(x=tips.total_bill, y = tips.tip, data = tips, hue = 'smoker')
```

```python
# determine frequency of data values in size column
tips['size'].value_counts()
```

```python
# group based on size
# color coding
sns.relplot(x=tips.total_bill, y = tips.tip, data = tips, hue = 'size')
```

```python
# group based on size
# shape coding
sns.relplot(x=tips.total_bill, y = tips.tip, data = tips, style = 'size')
```

```python
tips.head()
```

```python
# group based on size and time
# shape coding - size
# color coding - time
sns.relplot(x=tips.total_bill, y = tips.tip, data = tips, style = 'size', hue = 'time')
```

```python
# day vs total_bill
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# smoker vs tip
sns.catplot(x = 'smoker', y = 'tip', data = tips)
```

```python
# it is combination of scatter plot and histogram 
sns.jointplot(x = tips.total_bill, y = tips.tip, data = tips)
```

```python
# it is plot of all pairs of numerical data set in columns our data frame
sns.pairplot(tips)
```
