```python
# we need visualization becuse it is difficult to derive a sentiment from numbers
# with the help of visualization analysis becomes easy
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
print(x)
print(y)
```

```python
# step 1 -> making pairs of elements in x, y on the basis of index
# step 2 -> plotting on graph
plt.scatter(x, y)
```

```python
# Beautification
plt.scatter(x, y, c = 'r', alpha=0.3, marker="p") # you can keep any color code
# giving references to axes
plt.xlabel("X array")
plt.ylabel("Y array")
# title
plt.title("X array vs Y array Graph")
# adding grid
plt.grid()
# controlling figure size
plt.figure(figsize = (3, 2))
```

```python
x = np.linspace(1, 10, 100)
y = np.sin(x)
print(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

```python
plt.figure(figsize=(8, 2))
plt.plot(x, y, "--r")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sine function")
plt.grid()
```

```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)
print(x)
```

```python
print(y)
```

```python
# bar graph
plt.figure(figsize = (5, 5))
plt.bar(x, y, color = '#AF04F9')
plt.title("Bar graph")
plt.xlabel("Representing categorical data")
plt.ylabel("Representing numerical data")
```

```python
# horizontal bar graph
plt.figure(figsize = (5, 5))
plt.barh(x, y, color = '#AF04F9')
plt.title("Horizontal Bar graph")
plt.xlabel("Representing numerical data")
plt.ylabel("Representing categorical data")
```

```python
data = [1, 2, 3, 4, 5, 6, 5, 6, 6, 7, 7, 8, 8, 8, 8, 8]
# frequency of data
plt.hist(data, color="#57F904")
plt.xlabel("Data values")
plt.ylabel("Frequency")
plt.title("Histogram")
# data values on x axis
# frequency on y axis
```

```python
# 3 D graph
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z, c= 'r', alpha = 0.8)
plt.xlabel("X array")
plt.ylabel("Y array")
plt.show()
```

```python
ax.plot3D(x, y, z)
```
