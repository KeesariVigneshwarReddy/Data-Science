```python
# merge and join operation
import pandas as pd
data1 = pd.DataFrame({
                        "key1" : [1, 2, 4, 5, 6],
                        "key2" : [4, 5, 6, 7, 8],
                        "key3" : [3, 4, 5, 6, 6]
                    })
data2 = pd.DataFrame({
                        "key1" : [1, 2, 45, 6, 67],
                        "key4" : [56, 5, 6, 7, 8],
                        "key5" : [3, 56, 5, 6, 6]
                    })
print(data1)
print(data2)
```

```python
import pandas as pd
# intersection
pd.merge(data1, data2)
```

```python
pd.merge(data1, data2, how="left")
```

```python
pd.merge(data1, data2, how="right")
```

```python
pd.merge(data1, data2, how = "outer")
```

```python
pd.merge(data1, data2, how="cross")
```

```python
# it is same as default left
pd.merge(data1, data2, how="left", on = "key1")
```

```python
pd.merge(data1, data2, how="left", left_on = "key2", right_on = "key4")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

```python
"""
Questions
Q1) create a new column, convert Fare to INR
Q2) create a new column, which contain length of all names
Q3) create a new column, which describes range of passengers 0 - 10 -> cheap
                                                             10 - 20 -> mid
                                                             20+     -> high
"""
```

```python
# apply is a map function. We write a function in bracket and that will be implemented in each and every value
df["Fare_INR"] = df["Fare"].apply(lambda x : x * 89.62)
```

```python
df
```

```python
df["length_of_name"] = df["Name"].apply(len)
```

```python
df.head()
```

```python
def returnRange(x) :
    if x >= 0 and x < 10 :
        return "Cheap"
    elif x >= 10 and x < 20 :
        return "Mid"
    else : 
        return "High"

df["Range"] = df["Fare"].apply(returnRange)
```
