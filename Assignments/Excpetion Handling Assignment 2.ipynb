Arithmetic exception occurs when there is a wrong or invalid mathematical operation in our program.
1. Overflow error
This occurs when the number is too large and cannot stored in memory of system.
2. Zero division error
dividing a finite number with zero is an invalid operation in mathematics

```python
i = 1
try :
    f = 3.0**i
    for i in range(100):
        print (i, f)
        f = f ** 2
except OverflowError as e:
    print(e)
```

```python
try :
    a = 10 / 0
except ZeroDivisionError as e:
    print(e)
```
