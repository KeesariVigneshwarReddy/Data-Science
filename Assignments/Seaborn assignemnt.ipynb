# Question 1
Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Scatter plot

It is used to check out how the a feature x is scattered over with another feature y.
2. Line plot

It is used to understand how the value of feature x is changing with respect to feature y.
3. Bar plot

In a bar plot, categorical values are represented on x axis and its value is a bar. The height of bar is proportional to occurence of that category in column.
4. Histogram

It is used to plot frequency of elements (may be categorical or continous).
5. Heat map

Used for correlation plot of features

# Question 2
Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn')
df = sns.load_dataset("fmri")
```

```python
df.head()
```

```python
sns.lineplot(x = df['timepoint'], y = df['signal'])
plt.title("Line plot")
```

# Question 3
Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1)

sns.boxplot(x = df['pclass'], y = df['age'], ax = axes[0])
axes[0].set_title("Box plot 1")

sns.boxplot(x = df['pclass'], y = df['fare'], ax = axes[1])
axes[1].set_title("Box plot 2")
```

```python
sns.boxplot(x = df['pclass'], y = df['age'])
```

```python
sns.boxplot(x = df['pclass'], y = df['fare'])
```

# Question 4
Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
df = sns.load_dataset("diamonds")
df.head()
```

```python
sns.histplot(data = df, x = df['price'], kde = True, hue = 'cut', multiple = 'stack')
plt.title("Histogram")
```

# Question 5
Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
df = sns.load_dataset("iris")
sns.pairplot(data = df, hue = "species")
plt.title("Pair plot of iris dataset")
```

# Question 6
Use the "flights" dataset from seaborn to plot a heatmap.

```python
df = sns.load_dataset("flights")
df_rec = df.drop(columns=['month'])
sns.heatmap(df_rec, annot = True, cmap = 'coolwarm')
plt.title("Heat map of flights dataset")
```
