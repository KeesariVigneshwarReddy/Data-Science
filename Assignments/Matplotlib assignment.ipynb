# Question 2
What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x, y)
plt.title("Scatter plot")
plt.xlabel("x")
plt.ylabel("y")
```

# Question 3
Why is the subplot() function used? Draw four line plots using the subplot() function.

In a single figure, if we want to view multiple plots then we use subplot.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 1, figsize = (20, 10))

axes[0].plot(x1, y1)
axes[0].set_title("Line plot 1")
axes[0].set_xlabel("x1")
axes[0].set_ylabel("y1")

axes[1].plot(x2, y2)
axes[1].set_title("Line plot 2")
axes[1].set_xlabel("x2")
axes[1].set_ylabel("y2")

axes[2].plot(x3, y3)
axes[2].set_title("Line plot 3")
axes[2].set_xlabel("x3")
axes[2].set_ylabel("y3")

axes[3].plot(x4, y4)
axes[3].set_title("Line plot 4")
axes[3].set_xlabel("x4")
axes[3].set_ylabel("y4")
```

# Question 4
 What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

A bar plot represents categorical data with bars. Bar height is proportional to frequency of that category.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.title("Bar graph")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

# Question 5
What is a box plot? Why is it used? Using the following data plot a box plot.

Box plot is plot to detect outlier of a random variable. It represents minimum, Q1, Q2, Q3, maximum on the plot. Any value of random variable which lies outside the fence is an outlier


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].boxplot(box1)
axes[0].set_title("Box plot 1")
axes[0].set_xlabel("values")

axes[1].boxplot(box2)
axes[1].set_title("Box plot 2")
axes[1].set_xlabel("values")
```
