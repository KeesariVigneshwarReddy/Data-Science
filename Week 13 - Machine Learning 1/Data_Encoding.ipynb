Data Encoding is converting categorical features into numeric features. A ML model can only understand numeric values. 

### Nominal/OHE Encoding
One hot encoding, also known as nominal encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red, green, blue), we can represent it using one hot encoding as follows:

1. Red: [1, 0, 0]
2. Green: [0, 1, 0]
3. Blue: [0, 0, 1]

In a categorical feature if we have 3 categories, then 3 new columns will be created to reoresent the binary vector.

But if feature has 100 categories, then using one hot encoder is not a good idea.

One hot encoding creates a sparse matrix in dataframe. The model may get overfitted.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a simple dataframe
df = pd.DataFrame({
                      'color' : ['red', 'blue', 'green', 'red', 'blue']
                  })
df.head()
```

```python
# Create an instance of one hot encoder
encoder = OneHotEncoder()
# Fit and transform
encoder.fit_transform(df[['color']]).toarray()
```

Feature 1 = blue

Feature 2 = green

Feature 3 = red

Categories are in alphabetical order

```python
# creating a data frame
encoded_df = pd.DataFrame(encoder.fit_transform(df[['color']]).toarray(), columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
# concat original dataframe and encoded dataframe
pd.concat([df, encoded_df], axis=1)
```

# Internal Assignment

Perform one hot encoding with 'sex', 'smoker', 'day', 'time'

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
# Create an instance of one hot encoder
encoder = OneHotEncoder()
# Fit and transform
encoder.fit_transform(df[['sex', 'smoker', 'day', 'time']]).toarray()
```

```python
# creating a data frame
encoded_df = pd.DataFrame(encoder.fit_transform(df[['sex', 'smoker', 'day', 'time']]).toarray(), columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
# concat original dataframe and encoded dataframe
pd.concat([df, encoded_df], axis=1)
```
